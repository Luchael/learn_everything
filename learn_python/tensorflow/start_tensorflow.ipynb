```python
iris = np.loadtxt('iris.csv', skiprows=1, delimiter=',',usecols=[0,1,2,3])
X = iris[:,:3]
y = iris[:, [3]]

W = tf.Variable(tf.random.normal([3,1])) #가중치
b = tf.Variable(0.0) #절편값

lr = 0.001 #학습률
losses = [] #오차값
```

```python
for epoch in range(1000): # 1000번 반복
    with tf.GradientTape() as tape:
        pred_y = X@W+b                                   #출력값 계산
        loss = tf.reduce_sum((y - pred_y)**2)/len(y)     #MSE
        
    grads = tape.gradient(loss, [W, b])  #기울기 계산
    W.assign_sub(lr*grads[0])           #가중치값 변경
    b.assign_sub(lr*grads[1])
    
    losses.append(loss.numpy())
    
    if(epoch+1)%100 == 0:#그래프 그리기
        print(epoch, loss.numpy())
        plt.figure()
        plt.plot(losses)
        plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([3,5,9])
```

```python
model = Sequential()
model.add(Dense(1, input_shape = (1,), activation = 'linear')) # 'linear' is default
```

```python
model.compile(loss = 'mse', optimizer = 'sgd') # mse = 평균제곱오차 sgd = 경사하강법
model.fit(x.reshape(-1, 1), y, epochs = 1000)
```
